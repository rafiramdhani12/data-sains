```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

df_students = pd.read_csv("grades.csv" , delimiter=","  , header="infer")

df_students = df_students.dropna(axis=0 , how='any')

passes = pd.Series(df_students.Grade >= 60)
df_students = pd.concat([df_students , passes.rename("Pass")] , axis=1)

df_students
```

```python
# buat grafik batang 

fig = plt.figure(figsize=(12.8,7.2))

plt.bar(x=df_students.Name , height=df_students.Grade , color='orange')
plt.title("student grades")
plt.xlabel("Name")
plt.ylabel("Grade")
plt.xticks(rotation=90)

plt.show()
```

```python
fig , ax = plt.subplots(1,2,figsize=(10,4))

ax[0].bar(x=df_students.Name , height=df_students.Grade , color='orange')

ax[0].set_title("student grades")
ax[0].set_xticklabels(df_students.Name,rotation=90)

passes_count = df_students.Pass.value_counts()
ax[1].pie(passes_count , labels=passes_count)
ax[1].set_title("passing grades")

fig.show()
```

```python
# the fucking simple way bhow to create a chart
df_students.plot.bar(x="Name" , y="Grade" , figsize=(10,4), color="teal")
```

```python
grades = df_students.Grade

min_val = df_students.Grade.min()
max_val = df_students.Grade.max()
mean_val = df_students.Grade.mean()
mad_val = df_students.Grade.median()
mode_val = df_students.Grade.mode()[0]

fig = plt.figure(figsize=(12.8,7.2))

plt.hist(grades)

plt.axvline(x=min_val , color="gray" , label="min")
plt.axvline(x=max_val , color="gray" , label="max")
plt.axvline(x=mean_val , color="cyan" , label="mean")
plt.axvline(x=mad_val , color="red" , label="median")
plt.axvline(x=mode_val , color="yellow" , label="mode")


plt.title("data distribution")
plt.xlabel("grade")
plt.ylabel("count")

plt.show()
```

```python
grades = df_students.Grade

fig = plt.figure(figsize=(12.8,7.2))

plt.boxplot(grades)

plt.title("data distribution")
plt.xlabel("grade")
plt.ylabel("count")

plt.show()
```

```python
def show_density(data):
    fig = plt.figure(figsize=(10,4))
    
    data.plot.density()
    
    plt.show()
show_density(df_students.Grade)
```
